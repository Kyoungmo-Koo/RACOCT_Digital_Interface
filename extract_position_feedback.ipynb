```python
def compute_mae_rmse(actual_data_x, target_data_x, shift):
    # Ensure the shift is within the valid range
    if shift < 0 or shift >= len(actual_data_x):
        raise ValueError("Shift must be between 0 and the length of the data minus 1.")

    # Calculate the Mean Absolute Error (MAE)
    mae = np.mean(np.abs(actual_data_x[shift:] - target_data_x[:-shift]))

    # Calculate Mean Squared Error (MSE)
    mse = np.mean((actual_data_x[shift:] - target_data_x[:-shift]) ** 2)

    # Calculate Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)

    return mae, rmse
```

```python
# Range of shift values
shifts = range(10, 31)

# Initialize lists to store MAE and RMSE values
mae_values = []
rmse_values = []

# Compute MAE and RMSE for each shift value
for shift in shifts:
    mae, rmse = compute_mae_rmse(actual_data_x, target_data_x, shift)
    mae_values.append(mae)
    rmse_values.append(rmse)
    print(f"Shift: {shift} -> MAE: {mae}, RMSE: {rmse}")

# Plot MAE and RMSE against shift values
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(shifts, mae_values, marker='o', linestyle='-', color='b')
plt.title('Mean Absolute Error (MAE)')
plt.xlabel('Shift')
plt.ylabel('MAE')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(shifts, rmse_values, marker='o', linestyle='-', color='r')
plt.title('Root Mean Squared Error (RMSE)')
plt.xlabel('Shift')
plt.ylabel('RMSE')
plt.grid(True)

plt.tight_layout()
plt.show()
```
